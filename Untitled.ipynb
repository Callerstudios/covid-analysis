```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import dataset

```python
df = pd.read_csv("country_wise_latest.csv")
df.shape
```

```python
# Removing duplicates and empty values
new_df = df.drop_duplicates()
new_df = new_df.dropna()
new_df.shape
```

```python
deaths_df = new_df.loc[new_df["Deaths"] > (new_df["Deaths"].mean()), ["Country/Region", "Deaths"]]
deaths_df_sorted = deaths_df.sort_values(by="Deaths", ascending=False)
deaths_df_sorted.shape
```

```python
plt.figure(figsize=(50, 6))
sns.barplot(x="Country/Region", y="Deaths", data=deaths_df_sorted, width=1)
plt.xticks(rotation=90)
plt.show()
```

```python
correlation = new_df["Deaths"].corr(new_df["Recovered"])
correlation
```

```python
# Plotting correlation graph
plt.figure(figsize=(20, 6))
plt.scatter(new_df['Deaths'], new_df["Recovered"], color="blue")
plt.title("Correlation between deaths and Recoveries during COVID")
plt.xlabel("Deaths")
plt.ylabel("Recovered")
for i in range(len(new_df)):
    plt.annotate(new_df["Country/Region"][i], (new_df["Deaths"][i], new_df["Recovered"][i]), fontsize=9)
plt.grid()
plt.show()
```
